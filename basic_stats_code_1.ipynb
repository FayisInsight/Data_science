```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)  # Use sparse=False for newer versions

# Fit and transform the data
encoded_data = encoder.fit_transform(df[['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']])

# Convert the encoded data into a DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']))

# Convert float to int
encoded_df = encoded_df.astype(int)

# Concatenate with the original DataFrame (optional)
df_encoded = pd.concat([df, encoded_df], axis=1)

df_encoded.head()
```
